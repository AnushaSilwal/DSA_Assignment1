```python
class Node:
    def __init__(self, title):
        self.title = title
        self.next = None
        
```

```python
class Playlists:
    def __init__(self):
        self.head = None
    def add_song(self, title):
        new_node = Node(title)
        if  self.head is None:
            self.head = new_node
            print(f" '{title}' added to the playlist.")
            return
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node
        print(f" '{title}' added to the playlist.")
    def display_playlist(self):
        if self.head is None:
            print("The playlist is empty.")
            return
        print("current playlist:")
        current = self.head
        position = 1
        while current is not None:
            print(f"{position}. {current.title}")
            current = current.next
            position += 1
    def delete_song(self, title):
        if self.head is None:
            print("The playlist is empty.")
            return
        if self.head.title == title:
            self.head = self.head.next
            print(f" '{title}' removed from the playlist.")
            return
        current = self.head
        while current.next is not None:
            if current.next.title == title:
                current.next = current.next.next
                print(f" '{title}' removed from the playlist.")
                return
            current = current.next
        print(f" '{title}' not found in the playlist.")        
```

```python
playlist = Playlists()
playlist.add_song("blinding lights")
playlist.add_song("Shape of you") 
playlist.add_song("perfect")
playlist.display_playlist()
playlist.delete_song("shape of you")
playlist.display_playlist()
```
