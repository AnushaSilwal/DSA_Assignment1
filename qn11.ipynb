```python
class Player:
    def __init__(self, name):
        self.name = name
        self.next = None
```

```python
class circular_game:
    def __init__(self):
        self.head = None

    def add_player(self, name):
        new_player = Player(name)
        if self.head is None:
            self.head = new_player
            new_player.next = self.head
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = new_player
            new_player.next = self.head

    def show_turns(self, rounds):
        if self.head is None:
            return
        else:
            temp = self.head
            for _ in range(rounds):
                print(f"It's {temp.name}'s turn.")
                temp = temp.next
        
```

```python
game = circular_game()
game.add_player("Player1")
game.add_player("Player2")
game.add_player("Player3")      
game.add_player("Player4")
game.add_player("Player5")
game.show_turns(10)
```
