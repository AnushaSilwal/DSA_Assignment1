```python
class TextEditor:
    def __init__(self):
        self.text = ""
        self.history_stack = []

    def type_text(self, new_text):
        self.history_stack.append(self.text)
        self.text = new_text
        print(f"Typed: {new_text}")

    def undo(self):
        if self.history_stack:
            self.text = self.history_stack.pop()
            print(f"Undo: {self.text}")
        else:
            print("Nothing to undo.")

if __name__ == "__main__":
    editor = TextEditor()
    editor.type_text("A")
    editor.type_text("B.")
    editor.type_text("C")
    editor.undo()
    editor.undo()
```
