```python
class SongNode:
    def __init__(self, song):
        self.song = song    
        self.prev = None
        self.next = None
```

```python
class MusicPlayer:
    def __init__(self):
        self.current = None

    def add_song(self, song):
        new_song = SongNode(song)
        if self.current is None:
            self.current = new_song
            new_song.prev = new_song
            new_song.next = new_song
        else:
            last = self.current.prev
            last.next = new_song
            new_song.prev = last
            new_song.next = self.current
            self.current.prev = new_song

    def play_current(self):
            print(f" Now Playing: {self.current.song}")

    def previous_song(self):
            self.current = self.current.prev
            self.play_current()

    def next_song(self):
            self.current = self.current.next
            self.play_current()
        
        
```

```python
player = MusicPlayer()
player.add_song("Shape of You")
player.add_song("Believer")
player.add_song("Perfect")
player.play_current()  # Now Playing: Shape of You
player.next_song()     # Now Playing: Believer          
player.next_song()     # Now Playing: Perfect
player.next_song()    # Now Playing: Shape of You (circular)    
player.previous_song()  # Now Playing: Perfect

```
